```python
text = [
[1,2,3,4,5,6,7,8,9,1],
[2,3,4,5,6,7,8,9,1,2],
[3,4,5,6,7,8,9,1,2,3],
[4,5,6,7,8,9,1,2,3,4],
[5,6,7,8,9,1,2,3,4,5],
[6,7,8,9,1,2,3,4,5,6],
[7,8,9,1,2,3,4,5,6,7],
[8,9,1,2,3,4,5,6,7,8],
[9,1,2,3,4,5,6,7,8,9],
[1,2,3,4,5,6,7,8,9,1]
]

pattern = [
[8,9,1],
[9,1,2],
[1,2,3]
]

comparisons = 0
found = False

for i in range(8):
    for j in range(8):

        match = True

        for x in range(3):
            for y in range(3):

                comparisons += 1

                if text[i+x][j+y] != pattern[x][y]:
                    match = False
                    break

            if not match:
                break

        if match:
            print("Pattern found at:", (i, j))
            found = True
            break

    if found:
        break

print("Total Comparisons:", comparisons)
```
