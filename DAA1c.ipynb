```python
import random
import matplotlib.pyplot as plt

# Interpolation Search for Floating Point Numbers
def interpolation_search_float(arr, target):

    low = 0
    high = len(arr) - 1
    comparisons = 0

    while low <= high and arr[low] <= target <= arr[high]:

        comparisons += 1

        if arr[high] == arr[low]:
            break

        pos = int(
            low +
            ((target - arr[low]) * (high - low))
            / (arr[high] - arr[low])
        )

        if pos < low or pos > high:
            break

        if abs(arr[pos] - target) < 1e-6:
            return pos, comparisons

        elif arr[pos] < target:
            low = pos + 1

        else:
            high = pos - 1

    return -1, comparisons


sizes = [10000, 50000, 100000]
comparison_results = []

print("----- PERFORMANCE ANALYSIS -----")

for size in sizes:

    data = sorted(random.uniform(0.0, 1000.0) for _ in range(size))

    target = data[size // 2]

    index, comparisons = interpolation_search_float(data, target)

    comparison_results.append(comparisons)

    print(f"\nDataset Size = {size}")
    print("Target Found At Index =", index)
    print("Comparisons =", comparisons)

# Graph
plt.figure(figsize=(8,5))
plt.plot(sizes, comparison_results, marker='o')
plt.title("Interpolation Search Performance")
plt.xlabel("Dataset Size")
plt.ylabel("Number of Comparisons")
plt.grid(True)
plt.show()
```
